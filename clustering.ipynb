```python
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sb
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('star_with_gravity.csv')
df.head()
df.dtypes
```

```python
mass_list = df['Mass']
radius_list = df['Radius']
```

```python
l=[]
for i,m in enumerate(mass_list):
    if not (math.isnan(m) or math.isnan(radius_list[i])):
        tl=[radius_list[i],m]
        l.append(tl)
    # else:
        # print(m,radius_list[i])
# print(len(l))
```

```python
wcss = []
for i in range(1,11):
    t= KMeans(n_clusters=i,init='k-means++',random_state=23)
    t.fit(l)
wcss.append(t.inertia_)
wcss
```

```python
plt.figure(figsize=(10,5))
sb.lineplot(range(1,11) ,data= wcss,markers='o',color='orange')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
